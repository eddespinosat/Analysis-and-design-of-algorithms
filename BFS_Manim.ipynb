```python
def BFS(graph, start):
    discovered = {node: False for node in graph}
    discovered[start] = True
    layers = {0: [start]}
    bfs_tree = []
    i = 0

    while i in layers and layers[i]:
        layers[i + 1] = []
        for u in layers[i]:
            for v in graph[u]:
                if not discovered[v]:
                    discovered[v] = True
                    bfs_tree.append((u, v))
                    layers[i + 1].append(v)
        i += 1

    return bfs_tree, layers

# Ejemplo de cómo ingresar un grafo
graph_example = {
    1: [2, 3],
    2: [1, 4, 5],
    3: [1],
    4: [2],
    5: [2],
    6: [7, 8],
    7: [6],
    8: [6]
}

start_node = 5
bfs_tree, layers = BFS(graph_example, start_node)

# Imprimir los resultados
print("Árbol BFS:", bfs_tree)
print("Capas:", layers)
```

```python
%%manim -qm -v WARNING DrawGraph

from manim import *

class DrawGraph(Scene):
    def construct(self):
        # Definición del grafo
        graph_example = {
            1: [2, 3],
            2: [1, 4, 5],
            3: [1],
            4: [2],
            5: [2, 6],
            6: [5]
        }

        # Crear lista de aristas
        edges_list = [(u, v) for u, adj in graph_example.items() for v in adj if u < v]

        # Crear el objeto Graph con nodos y aristas
        G = Graph(list(graph_example.keys()), edges_list, layout="spring", layout_scale=3)

        # Crear objetos de nodo para la animación
        nodes = {node: Dot().move_to(G[node].get_center()) for node in graph_example}

        # Crear objetos de arista para la animación
        edges = {(u, v): Line(G[u].get_center(), G[v].get_center()) for u, adj in graph_example.items() for v in adj if u < v}

        # Algoritmo BFS con animación del nodo actual y resaltado de aristas
        def BFS_animation(graph, start, nodes, edges):
            discovered = {node: False for node in graph}
            discovered[start] = True
            layers = {0: [start]}
            bfs_tree = []
            i = 0

            while i in layers and layers[i]:
                layers[i + 1] = []
                for u in layers[i]:
                    self.play(nodes[u].animate.set_color(RED))
                    for v in graph[u]:
                        if not discovered[v]:
                            discovered[v] = True
                            bfs_tree.append((u, v))
                            layers[i + 1].append(v)
                            self.play(nodes[v].animate.set_color(GREEN))
                            # Resaltar la arista que se agrega al árbol BFS
                            edge = edges.get((u, v)) or edges.get((v, u))
                            if edge:
                                self.play(edge.animate.set_color(BLUE))
                i += 1

            return bfs_tree, layers

        # Crear etiquetas para los nodos con un tamaño de fuente más pequeño
        node_labels = {node: Text(str(node), font_size=24, color="YELLOW").next_to(G[node], DOWN, buff=0.1) for node in graph_example}

        # Animar la creación del grafo, etiquetas, nodos y aristas
        self.play(Create(G), *[Write(label) for label in node_labels.values()],
                  *[Create(nodes[node]) for node in nodes.keys()], *[Create(edges[edge]) for edge in edges.keys()])

        # Ejecutar BFS con animación
        start_node = 5  # Nodo de inicio para BFS
        bfs_tree, layers = BFS_animation(graph_example, start_node, nodes, edges)

        # Mantener la escena estática por unos segundos
        self.wait(3)
```

```python
%%manim -qm -v WARNING DrawGraph
config.media_width="80%"

from manim import *

class DrawGraph(Scene):
    def construct(self):
        # Definición del grafo
        graph_example = {
            1: [2, 3],
            2: [1, 3, 4, 5],
            3: [1, 2, 5, 7, 8],
            4: [2, 5],
            5: [2, 3, 4, 6],
            6: [5],
            7: [3, 8],
            8: [3, 7],
        }

        # Crear lista de aristas
        edges_list = [(u, v) for u, adj in graph_example.items() for v in adj if u < v]

        # Crear el objeto Graph con nodos y aristas
        G = Graph(list(graph_example.keys()), edges_list, layout="spring", layout_scale=3)

        # Crear objetos de nodo para la animación
        nodes = {node: Dot(radius=0.16).move_to(G[node].get_center()) for node in graph_example}

        # Crear objetos de arista para la animación
        edges = {(u, v): Line(G[u].get_center(), G[v].get_center(), stroke_width=7) for u, adj in graph_example.items() for v in adj if u < v}

        # Posición de la etiqueta de asignación de capas
        capa_label_pos = UP + LEFT

        mensaje_final_pos = DOWN + RIGHT

        # Algoritmo BFS con animación del nodo actual y resaltado de aristas
        def BFS_animation(graph, start, nodes, edges, capa_label_pos):
            discovered = {node: False for node in graph}
            discovered[start] = True
            layers = {0: [start]}
            bfs_tree = []
            i = 0

            while i in layers and layers[i]:
                layers[i + 1] = []
                for u in layers[i]:
                    # Animar el nodo actual con un efecto de destello
                    self.play(nodes[u].animate.set_color(RED))
                    self.play(nodes[u].animate.set_color(BLUE), run_time=0.3)  # Destello rápido
                    self.play(nodes[u].animate.set_color(RED), run_time=0.3)
                    self.play(nodes[u].animate.set_color(BLUE), run_time=0.3)  # Destello rápido
                    self.play(nodes[u].animate.set_color(RED))
                    for v in graph[u]:
                        if not discovered[v]:
                            discovered[v] = True
                            bfs_tree.append((u, v))
                            layers[i + 1].append(v)
                            # Animar el nodo descubierto
                            self.play(nodes[v].animate.set_color(GREEN))
                            self.play(nodes[v].animate.set_color(BLUE), run_time=0.3)  # Destello rápido
                            self.play(nodes[v].animate.set_color(GREEN), run_time=0.3)
                            self.play(nodes[v].animate.set_color(BLUE), run_time=0.3)  # Destello rápido
                            self.play(nodes[v].animate.set_color(GREEN))
                            # Mostrar la asignación de capa del nodo descubierto
                            capa_label = Text(f"Nodo {v} asignado a Capa {i+1}", font_size=20).to_edge(capa_label_pos)
                            self.play(FadeIn(capa_label))
                            self.wait(0.5)
                            self.play(FadeOut(capa_label))
                            # Resaltar la arista descubierta
                            edge = edges.get((u, v)) or edges.get((v, u))
                            if edge:
                                self.play(edge.animate.set_color(RED))
                                self.play(edge.animate.set_color(BLUE), run_time=0.3)
                                self.play(edge.animate.set_color(RED), run_time=0.3)
                                self.play(edge.animate.set_color(BLUE), run_time=0.3)
                                self.play(edge.animate.set_color(RED))
                i += 1

            return bfs_tree, layers

        # Crear etiquetas para los nodos con un tamaño de fuente más pequeño
        node_labels = {node: Text(str(node), font_size=24, color="YELLOW").next_to(nodes[node], DOWN, buff=0.1) for node in graph_example}

        # Animar la creación del grafo, etiquetas, nodos y aristas
        self.play(Create(G), *[Write(label) for label in node_labels.values()],
                  *[Create(nodes[node]) for node in nodes.keys()], *[Create(edges[edge]) for edge in edges.keys()])

        # Ejecutar BFS con animación incluyendo las asignaciones de capa
        start_node = 1  # Nodo de inicio para BFS
        bfs_tree, layers = BFS_animation(graph_example, start_node, nodes, edges, capa_label_pos)

        # Mostrar mensaje final
        mensaje_final = Text("La búsqueda por amplitud (BFS) ha finalizado", font_size=16).to_edge(mensaje_final_pos)
        self.play(FadeIn(mensaje_final))

        # Mantener la escena estática por 10 segundos
        self.wait(10)
```
