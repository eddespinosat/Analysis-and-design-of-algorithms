```python
def DFS_corrected(graph, start):
    explored = {node: False for node in graph}
    dfs_tree = []
    stack = [start]

    while stack:
        u = stack.pop()
        if not explored[u]:
            explored[u] = True
            for v in reversed(graph[u]):  # Se invierte para mantener el orden de la lista de adyacencia
                if not explored[v]:
                    dfs_tree.append((u, v))
                    stack.append(v)  # Se añade v a la pila para su posterior exploración

    return dfs_tree

# Definición del grafo con el orden de los nodos adyacentes tal como se muestra en la imagen
graph_example = {
    1: [2, 3],
    2: [1, 3],
    3: [1, 2, 5, 7, 8],
    4: [5],
    5: [3, 4, 6],
    6: [5],
    7: [3, 8],
    8: [3, 7]
}

# Nodo de inicio para DFS
start_node = 1

# Ejecución del DFS corregido
dfs_tree_corrected = DFS_corrected(graph_example, start_node)
```

```python
%%manim -qm -v WARNING DrawGraph

from manim import *

class DrawGraph(Scene):
    def construct(self):
        # Algoritmo DFS con animación del nodo actual
        def DFS_animation(graph, start):
            explored = {node: False for node in graph}
            dfs_tree = []
            stack = [start]

            while stack:
                u = stack.pop()
                if not explored[u]:
                    # Cambiar el color del nodo actual
                    self.play(nodes[u].animate.set_color(RED))
                    explored[u] = True
                    for v in reversed(graph[u]):
                        if not explored[v]:
                            dfs_tree.append((u, v))
                            stack.append(v)
                            # Cambiar el color del nodo descubierto
                            self.play(nodes[v].animate.set_color(GREEN))

            return dfs_tree

        # Definición del grafo
        graph_example = {
            1: [2, 3],
            2: [1, 4, 5],
            3: [1],
            4: [2],
            5: [2],
            6: [7, 8],
            7: [6],
            8: [6]
        }

        # Convertir el grafo a formato adecuado para Manim
        edges = [(u, v) for u, adj in graph_example.items() for v in adj if u < v]

        # Crear el objeto Graph
        G = Graph(list(graph_example.keys()), edges, layout="spring", layout_scale=3)

        # Crear etiquetas para los nodos con un tamaño de fuente más pequeño
        node_labels = {node: Text(str(node), font_size=24).next_to(G[node], DOWN, buff=0.1) for node in graph_example}

        # Crear objetos de nodo para la animación
        nodes = {node: Dot().move_to(G[node].get_center()) for node in graph_example}

        # Animar la creación del grafo, etiquetas y nodos
        self.play(Create(G), *[Write(label) for label in node_labels.values()], *[Create(dot) for dot in nodes.values()])

        # Ejecutar DFS con animación
        start_node = 1  # Nodo de inicio para DFS
        dfs_tree = DFS_animation(graph_example, start_node)

        # Mantener la escena estática por unos segundos
        self.wait(3)
```

```python
%%manim -qm -v WARNING DrawGraph
config.media_width = "80%"

class DrawGraph(Scene):
    def construct(self):
        # Definición del grafo
        graph_example = {
            1: [2, 3],
            2: [1, 3, 4, 5],
            3: [1, 2, 5, 7, 8],
            4: [2, 5],
            5: [2, 3, 4, 6],
            6: [5],
            7: [3, 8],
            8: [3, 7],
        }

        # Crear lista de aristas
        edges_list = [(u, v) for u, adj in graph_example.items() for v in adj if u < v]

        # Crear el objeto Graph con nodos y aristas
        G = Graph(list(graph_example.keys()), edges_list, layout="spring", layout_scale=3)

        # Crear objetos de nodo para la animación
        nodes = {node: Dot(radius=0.16).move_to(G[node].get_center()) for node in graph_example}

        # Crear objetos de arista para la animación
        edges = {(u, v): Line(G[u].get_center(), G[v].get_center(), stroke_width=7) for u, adj in graph_example.items() for v in adj if u < v}

        # Posición del mensaje de profundización
        profundizacion_label_pos = UP + LEFT

        mensaje_final_pos = DOWN + RIGHT

        # Algoritmo DFS con animación del nodo actual y resaltado de aristas
        def DFS_animation(graph, start, nodes, edges, profundizacion_label_pos):
            explored = {node: False for node in graph}
            dfs_tree = []
            stack = [(start, None)]  # Incluimos el nodo y el nodo padre en la pila

            while stack:
                u, parent = stack.pop()
                if not explored[u]:
                    explored[u] = True
                    # Animar el nodo actual con un efecto de destello
                    self.play(nodes[u].animate.set_color(RED))
                    if parent is not None:  # Si u no es el nodo raíz
                        # Resaltar la arista que conecta u con su nodo padre
                        edge = edges.get((parent, u)) or edges.get((u, parent))
                        self.play(edge.animate.set_color(RED))
                        # Mostrar mensaje de profundización
                        profundizacion_label = Text(f"Nodo {u} explorado en profundidad desde {parent}", font_size=20).to_edge(profundizacion_label_pos)
                        self.play(FadeIn(profundizacion_label))
                        self.wait(0.5)
                        self.play(FadeOut(profundizacion_label))
                        self.play(edge.animate.set_color(BLUE))
                    for v in reversed(graph[u]):  # Se invierte para mantener el orden de la lista de adyacencia
                        if not explored[v]:
                            dfs_tree.append((u, v))
                            stack.append((v, u))  # Añadir v a la pila con u como su padre
                            # Animar el nodo descubierto
                            self.play(nodes[v].animate.set_color(GREEN))

            return dfs_tree

        # Crear etiquetas para los nodos con un tamaño de fuente más pequeño
        node_labels = {node: Text(str(node), font_size=24, color="YELLOW").next_to(nodes[node], DOWN, buff=0.1) for node in graph_example}

        # Animar la creación del grafo, etiquetas, nodos y aristas
        self.play(Create(G), *[Write(label) for label in node_labels.values()],
                  *[Create(nodes[node]) for node in nodes.keys()], *[Create(edges[edge]) for edge in edges.keys()])

        # Ejecutar DFS con animación incluyendo mensajes de profundización
        start_node = 1  # Nodo de inicio para DFS
        dfs_tree = DFS_animation(graph_example, start_node, nodes, edges, profundizacion_label_pos)

        # Mostrar mensaje final
        mensaje_final = Text("La búsqueda en profundidad (DFS) ha finalizado", font_size=16).to_edge(mensaje_final_pos)
        self.play(FadeIn(mensaje_final))

        # Mantener la escena estática por 10 segundos
        self.wait(10)
```

```python
%%manim -qm -v WARNING DrawGraph
config.media_width = "80%"

from manim import *

class DrawGraph(Scene):
    def construct(self):
        # Definición del grafo
        graph_example = {
            1: [2, 3],
            2: [1, 3, 4, 5],
            3: [1, 2, 5, 7, 8],
            4: [2, 5],
            5: [2, 3, 4, 6],
            6: [5],
            7: [3, 8],
            8: [3, 7],
        }

        # Crear lista de aristas
        edges_list = [(u, v) for u, adj in graph_example.items() for v in adj if u < v]

        # Crear el objeto Graph con nodos y aristas
        G = Graph(list(graph_example.keys()), edges_list, layout="spring", layout_scale=3)

        # Crear objetos de nodo para la animación
        nodes = {node: Dot(radius=0.16).move_to(G[node].get_center()) for node in graph_example}

        # Crear objetos de arista para la animación
        edges = {(u, v): Line(G[u].get_center(), G[v].get_center(), stroke_width=7) for u, adj in graph_example.items() for v in adj if u < v}

        # Posición del mensaje de profundización y del mensaje final
        profundizacion_label_pos = UP + LEFT
        mensaje_final_pos = DOWN + RIGHT

        # Algoritmo DFS con animación del nodo actual y resaltado de aristas
        def DFS_animation(graph, start, nodes, edges, profundizacion_label_pos):
            explored = {node: False for node in graph}
            dfs_tree = []
            stack = [(start, None)]  # Incluimos el nodo y el nodo padre en la pila

            while stack:
                u, parent = stack.pop()
                if not explored[u]:
                    explored[u] = True
                    self.play(nodes[u].animate.set_color(RED))
                    if parent is not None:
                        edge = edges.get((parent, u)) or edges.get((u, parent))
                        if edge:
                            self.play(edge.animate.set_color(RED))
                    for v in reversed(graph[u]):
                        if not explored[v]:
                            dfs_tree.append((u, v))
                            stack.append((v, u))
                            self.play(nodes[v].animate.set_color(GREEN))
                    if parent is not None:
                        # Comprobar si todos los hijos han sido explorados para animar el retroceso
                        if all(explored[v] for v in graph[u]):
                            edge = edges.get((parent, u)) or edges.get((u, parent))
                            if edge:
                                self.play(Flash(edge.get_center(), color=WHITE, line_length=0.2))
                                self.play(edge.animate.set_color(ORANGE))

            return dfs_tree

        # Crear etiquetas para los nodos con un tamaño de fuente más pequeño
        node_labels = {node: Text(str(node), font_size=24, color="YELLOW").next_to(nodes[node], DOWN, buff=0.1) for node in graph_example}

        # Animar la creación del grafo, etiquetas, nodos y aristas
        self.play(Create(G), *[Write(label) for label in node_labels.values()],
                  *[Create(nodes[node]) for node in nodes.keys()], *[Create(edges[edge]) for edge in edges.keys()])

        # Ejecutar DFS con animación incluyendo mensajes de profundización y efectos de retroceso
        start_node = 1
        dfs_tree = DFS_animation(graph_example, start_node, nodes, edges, profundizacion_label_pos)

        # Mostrar mensaje final
        mensaje_final = Text("La búsqueda en profundidad (DFS) ha finalizado", font_size=16).to_edge(mensaje_final_pos)
        self.play(FadeIn(mensaje_final))

        # Mantener la escena estática por 10 segundos
        self.wait(10)
```
